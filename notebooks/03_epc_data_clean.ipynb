```python
import os
import io
import base64
from pathlib import Path

import requests
import pandas as pd
from dotenv import load_dotenv
```

```python
PROJECT_ROOT = Path(r"C:\Users\swapn\OneDrive\Desktop\fuel_poverty_risk_england")
RAW_PATH = PROJECT_ROOT / "data" / "raw"
RAW_PATH.mkdir(parents=True, exist_ok=True)

load_dotenv(PROJECT_ROOT / ".env")

EPC_EMAIL = os.getenv("EPC_EMAIL")
EPC_API_KEY = os.getenv("EPC_API_KEY")

LA_CODE = "E06000047"   # County Durham
BASE_URL = "https://epc.opendatacommunities.org/api/v1/domestic/search"

assert EPC_EMAIL, "EPC_EMAIL missing in .env"
assert EPC_API_KEY, "EPC_API_KEY missing in .env"

print("Paths and environment loaded successfully")
```

```python
credentials = f"{EPC_EMAIL}:{EPC_API_KEY}"
token = base64.b64encode(credentials.encode("utf-8")).decode("utf-8")

HEADERS = {
    "Accept": "text/csv",
    "Authorization": f"Basic {token}"
}

print("Authorization header prepared")
```

```python
params = {
    "local-authority": LA_CODE,
    "size": 200
}

resp = requests.get(BASE_URL, headers=HEADERS, params=params, timeout=60)

print("Status code:", resp.status_code)
print("Content-Type:", resp.headers.get("Content-Type"))
print("X-Next-Search-After:", resp.headers.get("X-Next-Search-After"))

if resp.status_code != 200:
    print(resp.text[:500])
else:
    epc_test = pd.read_csv(io.StringIO(resp.text))
    print("Shape:", epc_test.shape)
    display(epc_test.head())
```

```python
def fetch_epc_local_authority(local_authority_code, page_size=5000, max_pages=None):
    if page_size > 5000:
        raise ValueError("page_size cannot exceed 5000")

    frames = []
    search_after = None
    page = 1

    while True:
        params = {
            "local-authority": local_authority_code,
            "size": page_size
        }

        if search_after:
            params["search-after"] = search_after

        resp = requests.get(BASE_URL, headers=HEADERS, params=params, timeout=120)

        if resp.status_code != 200:
            raise RuntimeError(
                f"Request failed on page {page}: {resp.status_code}\n{resp.text[:500]}"
            )

        df = pd.read_csv(io.StringIO(resp.text))
        print(f"Page {page}: {df.shape[0]} rows")

        if not df.empty:
            frames.append(df)

        search_after = resp.headers.get("X-Next-Search-After")

        if not search_after:
            break

        if max_pages is not None and page >= max_pages:
            break

        page += 1

    if not frames:
        return pd.DataFrame()

    out = pd.concat(frames, ignore_index=True)

    if "lmk-key" in out.columns:
        out = out.drop_duplicates(subset=["lmk-key"], keep="first")

    return out
```

```python
epc_latest["is_low_rating"] = epc_latest["current-energy-rating"].isin(["E", "F", "G"]).astype(int)
epc_latest["is_low_efficiency"] = (epc_latest["current-energy-efficiency"] < 55).astype(int)
epc_latest["improvement_gap"] = (
    epc_latest["potential-energy-efficiency"] - epc_latest["current-energy-efficiency"]
)

print(epc_latest[["is_low_rating", "is_low_efficiency", "improvement_gap"]].describe())
print(epc_latest["is_low_rating"].value_counts())
print(epc_latest["is_low_efficiency"].value_counts())
```

```python
epc_latest["improvement_gap_clean"] = epc_latest["improvement_gap"].where(
    (epc_latest["improvement_gap"] >= 0) & (epc_latest["improvement_gap"] <= 100)
)

epc_latest["improvement_gap_clean"].describe()
```

```python
latest_file = RAW_PATH / f"epc_latest_engineered_{LA_CODE}.csv"
epc_latest.to_csv(latest_file, index=False)
print("Saved:", latest_file)
```

```python
postcode_risk = (
    epc_latest
    .groupby("postcode", dropna=False)
    .agg(
        properties=("building-reference-number", "count"),
        share_low_rating=("is_low_rating", "mean"),
        share_low_efficiency=("is_low_efficiency", "mean"),
        avg_current_efficiency=("current-energy-efficiency", "mean"),
        avg_potential_efficiency=("potential-energy-efficiency", "mean"),
        avg_improvement_gap=("improvement_gap_clean", "mean")
    )
    .reset_index()
)

postcode_risk.head()
```

```python
postcode_file = RAW_PATH / f"postcode_risk_{LA_CODE}.csv"
postcode_risk.to_csv(postcode_file, index=False)
print("Saved:", postcode_file)
```

```python
postcode_risk.sort_values(
    ["share_low_rating", "share_low_efficiency", "properties"],
    ascending=[False, False, False]
).head(20)
```

```python
postcode_risk_filtered = postcode_risk[postcode_risk["properties"] >= 20].copy()

postcode_risk_filtered.sort_values(
    ["share_low_rating", "share_low_efficiency", "properties"],
    ascending=[False, False, False]
).head(20)
```

```python
postcode_risk_filtered["risk_score"] = (
    0.4 * postcode_risk_filtered["share_low_rating"] +
    0.4 * postcode_risk_filtered["share_low_efficiency"] +
    0.2 * (postcode_risk_filtered["avg_improvement_gap"] / postcode_risk_filtered["avg_improvement_gap"].max())
)

postcode_risk_filtered.sort_values("risk_score", ascending=False).head(20)
```

```python
postcode_score_file = RAW_PATH / f"postcode_risk_scored_{LA_CODE}.csv"
postcode_risk_filtered.to_csv(postcode_score_file, index=False)
print("Saved:", postcode_score_file)
```

```python
postcode_risk_filtered["risk_band"] = pd.qcut(
    postcode_risk_filtered["risk_score"],
    q=5,
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

postcode_risk_filtered[["postcode", "properties", "risk_score", "risk_band"]].sort_values(
    "risk_score", ascending=False
).head(20)
```

```python
final_file = RAW_PATH / f"postcode_risk_final_{LA_CODE}.csv"
postcode_risk_filtered.to_csv(final_file, index=False)
print("Saved:", final_file)
```
