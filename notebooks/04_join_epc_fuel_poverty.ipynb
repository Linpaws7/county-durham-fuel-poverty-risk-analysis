```python
fuel_poverty_la.columns = fuel_poverty_la.iloc[0]
fuel_poverty_la = fuel_poverty_la.iloc[1:].reset_index(drop=True)
fuel_poverty_la.head()
```

```python
fuel_poverty_la_cd = fuel_poverty_la[
    fuel_poverty_la["Local Authority code"] == LA_CODE
].copy()

fuel_poverty_la_cd
```

```python
postcode_join = epc_postcode.copy()
postcode_join["la_fuel_poverty_pct"] = float(
    fuel_poverty_la_cd["Proportion of households fuel poor (%)"].iloc[0]
)
postcode_join.head()
```

```python
join_file = RAW_PATH / f"postcode_epc_fuelpoverty_{LA_CODE}.csv"
postcode_join.to_csv(join_file, index=False)
print("Saved:", join_file)
```
