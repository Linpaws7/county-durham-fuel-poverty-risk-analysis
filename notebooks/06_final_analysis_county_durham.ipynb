```python
df.sort_values("risk_score", ascending=True)[
    ["postcode", "properties", "risk_score", "risk_band", "share_low_rating", "avg_current_efficiency", "avg_improvement_gap"]
].head(20)
```

```python
summary = pd.DataFrame([{
    "local_authority": LA_CODE,
    "postcode_count": len(df),
    "mean_risk_score": df["risk_score"].mean(),
    "median_risk_score": df["risk_score"].median(),
    "highest_risk_postcode": df.sort_values("risk_score", ascending=False)["postcode"].iloc[0],
    "lowest_risk_postcode": df.sort_values("risk_score", ascending=True)["postcode"].iloc[0],
    "mean_current_efficiency": df["avg_current_efficiency"].mean(),
    "mean_potential_efficiency": df["avg_potential_efficiency"].mean(),
    "mean_improvement_gap": df["avg_improvement_gap"].mean()
}])

summary
```

```python
summary.to_csv(RAW_PATH / f"final_summary_{LA_CODE}.csv", index=False)
print("saved")
```
